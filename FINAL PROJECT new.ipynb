PROJECT TITLE: MUTUAL FUNDS REPORT ANALYSIS

INTRODUCTION
This project analyzes MUTUAL FUNDS REPORT DATASET obtained from Data.gov.in . The dataset contains detailed information about various mutual fund schemes, including assets under management, fund mobilization, redemptions, portfolio counts and scheme categories.
Through systematic data preprocessing and exploratory data analysis, the dataset was cleaned, transformed, and structured for meaningful analysis.
The analysis aims to extract meaningful insights and present a clear view of mutual fund behaviour over time.

ATTRIBUTE INFORMATION

The mutual funds report dataset consists of several attributes. These attributes are used to compare performance, assess risk, and analyze trends
across different mutual fund schemes.

Month : Reporting month and year of the mutual funds data.
Scheme Name: Name of the mutual fund scheme.
Scheme Type: Type of the mutual fund based on investment structure.
Scheme Category: Category of the mutual fund
Number of Schemes: Total number of mutual fund schemes available in the given period
Number of Portfolios: Total number of investor portfolios associated with the schemes
Funds Mobilized: Total amount of funds collected from investors 
Repurchase/Redemption: Amount withdrawn or redeemed by investors
Net Inflow/Outflow :  Net difference between funds mobilized and repurchased
Net Assets Under Management: Total market value of assets managed by the mutual funds
Average Net Assets Under Management:  Average value of assets managed over the reporting period
Number of Segregated Portfolios: Number of segregated portfolios created within the schemes
Net Assets Under Management in Segregated Portfolios: Total assets managed within segregated portfolios

PRE-PROCESSING

Data preprocessing is an essential step in data analytics that involves cleaning, transforming, and organizing the dataset by handling missing values, correcting data types, and removing inconsistencies to ensure accurate analysis.

1. Handling Missing Values

Checking missing values:
Missing values can affect overall accuracy of analysis. The 'df.isnull()' function identify missing values and the 'sum()' function count how many null values present in each column.

```python
# Checking missing values 
df.isnull().sum()
```

```python
# Filling Missing Values
df.fillna(0, inplace=True)

df.isnull().sum()
```

2. Correcting Data Types:
In this step, date columns and numerical attributes are converted into appropriate data types.   

```python
# Convert Month column to datetime
df['Month (date)'] = pd.to_datetime(df['Month (date)'])

df.info()
```

3. Rename Columns

```python
# Rename columns
df.rename(columns={
    'Month (date)': 'Date',
    'Scheme Name (scheme_name)': 'Scheme Name',
    'Scheme Type (scheme_type)': 'Scheme Type',
    'Scheme Category (scheme_category)': 'Scheme Category',
    'Number of Schemes (no_of_schemes)': 'Number of Schemes',
    'Number of Portfolios (no_of_folios)': 'Number of Portfolios',
    'Funds Mobilized (fund_mobilized)': 'Funds Mobilized',
    'Repurchase/Redemption (repurchase_redemption)': 'Repurchase/Redemption',
    'Net Inflow/Outflow (net_inflow_outflow)': 'Net Inflow/Outflow',
    'Net Assets Under Management (net_aum)': 'Net Assets Under Management(Net AUM)',
    'Average Net Assets Under Management (avg_net_aum)': 'Avg Net AUM',
    'Number of Segregated Portfolios (no_of_segregated_portfolios)': 'Number of Segregated Portfolios',
    'Net Assets Under Management in Segregates Portfolio (net_aum_in_segregated_portfolios)': 'Net AUM in Segregates Portfolio'
}, inplace=True)
```

```python
df.columns
```

4. Derived Columns
Derived columns used to add new columns to the dataset by using existing columns. These new columns provide additional insights and make analysis easier.

Derived Column for Flow Status(Inflow/ Outflow):
The derived column for flow status indicates whether a mutual fund experienced net inflow or net outflow during the reporting period.

```python
# Derived column for Flow Status(Inflow/Outflow)
df['flow_status'] = df['Net Inflow/Outflow'].apply(
    lambda x: 'Inflow' if x > 0 else 'Outflow'
)

# Display result
df[['Net Inflow/Outflow', 'flow_status']].head(10)
```

Derived Column for AUM(Asset Under Management) Difference:
This derived column captures the difference between actual net asset under management and average net assets under management.it helps to analyze asset variation.

```python
df['aum_difference'] = df['Net Assets Under Management(Net AUM)'] - df['Avg Net AUM']
df[['Net Assets Under Management(Net AUM)', 'Avg Net AUM', 'aum_difference']].head(12)
```

5. Filtering Data:
Filtering is used to select only the required rows based on specific conditions to focus on relevant data.

Filter Schemes with Positive Inflow:
This filtering step extracts schemes that experienced positive net fund inflows.

```python
# Filter Schemes with Positive Inflow
df[df['Net Inflow/Outflow'] > 0][
    ['Scheme Name', 'Net Inflow/Outflow', 'flow_status']
].head(15)
```

Filter data from Net AUM and Avg Net AUM for the year 2023:
In this step only the records from  Net AUM and Avg Net AUM for the year 2023 were filtered to analyze recent mutual fund performance.

```python
# Filter Net AUM & Avg Net AUM for 2023
#  Derived column for year is created 
df['year'] = df['Date'].dt.year

df_2023 = df[df['year'] == 2023][
    [
        'Net Assets Under Management(Net AUM)',
        'Avg Net AUM',
    ]
    ]   
df_2023.head(8)
```

6. Aggregating Data:
Aggregation means summarizing data using mathematical operations such as sum, mean, count, max/min.

Aggregate Net AUM by Year:
A year column was derived from the Date column to enable time-based analysis. The data was then grouped by year, and the total Net Assets Under Management (Net AUM) was calculated to study yearly asset trends.

```python
#  Derived column for year is created 
df['year'] = df['Date'].dt.year 
```

```python
# Aggregate Net AUM by Year
aum_by_year = df.groupby('year')['Net Assets Under Management(Net AUM)'].sum().reset_index()
aum_by_year
```

Total Funds Mobilized v/s Redemption:
This aggregation summarizes total funds mobilized and redeemed on a yearly basis.

```python
# Total Funds Mobilized v/s Redemption
fund_summary = df.groupby('year')[['Funds Mobilized', 'Repurchase/Redemption']].sum().reset_index()
fund_summary
```

Count of Schemes by Category: 
This step determines the number of unique mutual fund schemes in each category.

```python
# Count of Schemes by Category
scheme_count = df.groupby('Scheme Category')['Scheme Name'].nunique().reset_index()
scheme_count
```

 DATA VISUALIZATION

Data visualization helps transform numerical data into visual insights, making trends, patterns, and comparisons easy to understand. In this project, Matplotlib and Seaborn are used to create meaningful visualizations with proper titles, labels and legends. 

1. Bar Plot  Net AUM by Scheme Category:
This is used to compare how Net Assets Under Management (Net AUM) vary across different scheme categories.This bar chart shows which scheme category manages the highest total assets, helping identify dominant fund types.

```python
#  Bar Plot : Net AUM by Scheme Category:
plt.figure(figsize=(10,6))
sns.barplot(
    data=df,
    x='Scheme Category',
    y='Net Assets Under Management(Net AUM)',
    estimator=sum
)

plt.title('Total Net AUM by Scheme Category')
plt.xlabel('Scheme Category')
plt.ylabel('Total Net AUM')
plt.xticks(rotation=45)
plt.show()
```

2. Line Chart  Net AUM Trend Over Time:
It analyze how Net AUM changes over time. The line chart highlights growth or decline trends in asset management over different months.

```python
# Line Chart : Net AUM Trend Over Time
plt.figure(figsize=(10,6))
sns.lineplot(
    data=df,
    x='Date',
    y='Net Assets Under Management(Net AUM)'
)

plt.title('Net AUM Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Net AUM')
plt.show()
```

3. Pie Chart  Distribution of Scheme Categories:
It visualize the proportion of different scheme categories. This pie chart gives a quick overview of how mutual fund schemes are distributed across categories.

```python
# Pie Chart:  Distribution of Scheme Categories
category_counts = df['Scheme Category'].value_counts()

plt.figure(figsize=(8,8))
plt.pie(
    category_counts,
    labels=category_counts.index,
    autopct='%1.1f%%',
    startangle=140
)

plt.title('Scheme Category Distribution')
plt.show()
```

4. Histogram  Net Inflow/Outflow Distribution:
It used to understand the spread and frequency of inflows and outflows.The histogram shows whether inflows or outflows are more common and how they are distributed.

```python
# Histogram: Net Inflow/Outflow Distribution 
plt.figure(figsize=(10,6))
plt.hist(df['Net Inflow/Outflow'], bins=20)

plt.title('Distribution of Net Inflow/Outflow')
plt.xlabel('Net Inflow/Outflow')
plt.ylabel('Frequency')
plt.show()
```

5. Box Plot:
 A Box plot is a statistical visualization used to display the distribution of numerical data. Here, the box plot illustrates the distribution of Net Assets Under Management (Net AUM) across different years. Each box represents the spread of Net AUM values for a specific year. This visualization is useful for comparing year-wise fluctuations in Net AUM and understanding how consistently mutual fund assets have grown or varied across years.

```python
# Box Plot: Net AUM by Year
plt.figure(figsize=(10,6))
sns.boxplot(
    data=df,
    x='year',
    y='Net Assets Under Management(Net AUM)'
)

plt.title('Year-wise Net AUM Distribution')
plt.show()
```

6. Scatter Plot  Net AUM vs Net Inflow/Outflow:
Scatter plot is used to display the relationship between two numerical variables. In this analysis, scatter plot is used to examine the relationship between Net Inflow/Outflow and Net Assets Under Management (Net AUM). By plotting Net Inflow/Outflow on the x-axis and Net AUM on the y-axis, the visualization helps determine whether higher inflows are associated with higher asset value.

```python
# Scatter Plot: Net AUM v/s Net Inflow/Outflow
plt.figure(figsize=(10,6))
sns.scatterplot(
    data=df,
    x='Net Inflow/Outflow',
    y='Net Assets Under Management(Net AUM)',
    hue='Scheme Category'
)

plt.title('Net AUM vs Net Inflow/Outflow')
plt.show()
```

7. Heat Map - Correlation Between Numerical Variables:
A heatmap is a visual representation of data where values are shown using color intensity. In correlation analysis, a heatmap is used to display the strength and direction of relationships between numerical variables.  Correlation values range from -1 to +1, where values close to +1 indicate a strong positive relationship, values close to -1 indicate a strong negative relationship, and values around 0 indicate little or no correlation.

  In this analysis, the heatmap visualizes the correlations among key numerical variables such as Net Assets Under Management (Net AUM), Net Inflow/Outflow, Funds Mobilized, Repurchase/Redemption, and Average Net AUM. Warmer colors indicate stronger positive correlations, while cooler colors represent negative relationships.

```python
# Heat Map:  Correlation Between Numerical Variables
plt.figure(figsize=(10,6))
sns.heatmap(
    df.select_dtypes(include='number').corr(),
    annot=True,
    cmap='coolwarm'
)

plt.title('Correlation Heatmap')
plt.show()
```

8. Subplot - Multiple Visuals Together:
Subplots are used to display multiple visualizations within a single figure, allowing different insights to be compared side by side. 
In this analysis, two visualizations are presented together. The first subplot uses a bar chart to show the total Net Assets Under Management (Net AUM) across different scheme categories, highlighting which categories manage the highest total assets. The second subplot uses a box plot to display the distribution of Net Inflow/Outflow for the same scheme categories, revealing variability and the presence of inflows or outflows.
By placing these charts together, it becomes easier to compare asset size with flow behavior across categories. 

```python
# Subplots
fig, axes = plt.subplots(1, 2, figsize=(14,5))

# Bar plot: Total Net AUM by Category
sns.barplot(
    data=df,
    x='Scheme Category',
    y='Net Assets Under Management(Net AUM)',
    estimator=sum,
    ax=axes[0]
)
axes[0].set_title('Total Net AUM by Category')
axes[0].tick_params(axis='x', rotation=45)

# Box plot: Net Inflow/Outflow Distribution
sns.boxplot(
    data=df,
    x='Scheme Category',
    y='Net Inflow/Outflow',
    ax=axes[1]
)
axes[1].set_title('Net Inflow/Outflow Distribution by Category')
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
